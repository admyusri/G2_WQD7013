```python
# B2. Outlier Investigation
# Apply the IQR fence rule to at least two numerical variables.
# Apply the Z-score rule using |z| > 3 to the same variables.
# Count how many values each method flags and whether the flagged sets overlap.
# Investigate at least two flagged values in the raw data and decide whether
# they are plausible or likely errors.

# Choose two variables
var1 = "GDP_current_US"
var2 = "inflation_annual%"

# Import numpy for Z-score calculation
import numpy as np

# IQR rule for var1
col1 = df[var1].dropna()
Q1 = col1.quantile(0.25)
Q3 = col1.quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
iqr_outliers_var1 = col1[(col1 < lower) | (col1 > upper)]

# Z-score rule for var1
z_scores1 = (col1 - col1.mean()) / col1.std()
z_outliers_var1 = col1[np.abs(z_scores1) > 3]

print("Variable:", var1)
print("IQR outliers:", len(iqr_outliers_var1))
print("Z-score outliers:", len(z_outliers_var1))
print("Overlap:", len(set(iqr_outliers_var1.index) & set(z_outliers_var1.index)))
print("Flagged:", iqr_outliers_var1.head(2).values)

# Repeat for var2
col2 = df[var2].dropna()
Q1 = col2.quantile(0.25)
Q3 = col2.quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
iqr_outliers_var2 = col2[(col2 < lower) | (col2 > upper)]

z_scores2 = (col2 - col2.mean()) / col2.std()
z_outliers_var2 = col2[np.abs(z_scores2) > 3]

print("\nVariable:", var2)
print("IQR outliers:", len(iqr_outliers_var2))
print("Z-score outliers:", len(z_outliers_var2))
print("Overlap:", len(set(iqr_outliers_var2.index) & set(z_outliers_var2.index)))
print("Flagged:", iqr_outliers_var2.head(2).values)
```

## Justification of using .dropna() for both variables:
NaN values were excluded using .dropna() for 'GDP (current US)' and 'Inflation annual %' before applying the IQR and Z‑score rules. This ensures that outlier detection is based only on valid numeric observations, avoids distortions from artificial replacements and keeps the flagged values meaningful and interpretable.

```python
# Histogram of one numerical variable, with a justified bin count using either Sturges or Freedman-Diaconis rule

num_var = df.select_dtypes(include='number').columns[0]   # pick one numeric
cat_var = "date"   # good categorical variable
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = df[num_var].dropna()
n = len(data)

# Sturges
sturges_bins = int(np.ceil(1 + np.log2(n)))

# Freedman–Diaconis
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)
iqr = q3 - q1

fd_width = 2 * iqr / (n ** (1/3))
fd_bins = int(np.ceil((data.max() - data.min()) / fd_width))

print("n =", n)
print("Sturges bins =", sturges_bins)
print("FD bins =", fd_bins)

# Use FD (better)
plt.figure(figsize=(8,6))
plt.hist(data, bins=fd_bins, edgecolor="black")
plt.title(f"Histogram of {num_var}")
plt.xlabel(num_var)
plt.ylabel("Frequency")
plt.show()
```

```python
# A density-overlaid histogram with an approximate theoretical normal curve for the same variable.
import seaborn as sns
from scipy.stats import norm

mean = data.mean()
std = data.std()

x = np.linspace(data.min(), data.max(), 100)

plt.figure(figsize=(8,6))
sns.histplot(data, bins=fd_bins, stat="density", kde=True)

plt.plot(x, norm.pdf(x, mean, std))

plt.title("Density + Normal Curve")
plt.show()
```

### D1.4 Verifying Probabilities

#### D1.4.1 Verifying Bayes' Theorem
The successful verification of Bayes' Theorem demonstrates that conditional probabilities are systematically reversible. Our initial calculation tells us if a country becomes wealthy, it is highly likely to become a high corruption. However, Bayes' Theorem allows us to work backward from an observation. By calculation, we establish that corruption act as an important key for national wealth. If we observe a country with high corruption without knowing any of its economic data, we can use Bayes' Theorem to update our beliefs and predict that the country belongs to the upper half of global GDPs. This highlights that high corruption are almost exclusively a characteristic of wealthy nations in this dataset.

```python
# Bayes Formula: P(A|B) = [P(B|A) * P(A)] / P(B)
bayes_calculation1 = (p_high_corrupt_given_high_gdp * p_high_gdp) / p_high_corrupt
# Bayes Formula: P(A|B') = [P(B'|A) * P(A)] / P(B')
bayes_calculation2 = (p_low_corrupt_given_high_gdp * p_high_gdp) / p_low_corrupt
#P(B|A) = [P(A|B) * P(B)] / P(A)
bayes_calculation3 = (p_high_gdp_given_high_corrupt * p_high_corrupt) / p_high_gdp
#P(B|A') = [P(A'|B) * P(B)] / P(A')
bayes_calculation4 = (p_low_gdp_given_high_corrupt * p_high_corrupt) / p_low_gdp

print(f"Directly calculated P(A | B): {p_high_gdp_given_high_corrupt:.4f}")
print(f"Calculated via Bayes' Theorem: {bayes_calculation1:.4f}")
if round(p_high_gdp_given_high_corrupt, 4) == round(bayes_calculation1, 4):
    print("Result: Bayes' Theorem is verified!\n")

print(f"Directly calculated P(A | B'): {p_high_gdp_given_low_corrupt:.4f}")
print(f"Calculated via Bayes' Theorem: {bayes_calculation2:.4f}")
if round(p_high_gdp_given_low_corrupt, 4) == round(bayes_calculation2, 4):
    print("Result: Bayes' Theorem is verified!\n")

print(f"Directly calculated P(B | A): {p_high_corrupt_given_high_gdp:.4f}")
print(f"Calculated via Bayes' Theorem: {bayes_calculation3:.4f}")
if round(p_high_corrupt_given_high_gdp, 4) == round(bayes_calculation3, 4):
    print("Result: Bayes' Theorem is verified!\n")

print(f"Directly calculated P(B | A'): {p_high_corrupt_given_low_gdp:.4f}")
print(f"Calculated via Bayes' Theorem: {bayes_calculation4:.4f}")
if round(p_high_corrupt_given_low_gdp, 4) == round(bayes_calculation4, 4):
    print("Result: Bayes' Theorem is verified!\n")
```

## Task D2: Monte Carlo Simulation
Monte Carlo simulations are computational algorithms that rely on repeated random sampling to estimate numerical results. Instead of relying purely on theoretical mathematical formulas, Monte Carlo methods allow us to use code and randomness to approximate probabilities and model complex systems.



**D2a: Biased Coin Simulation**

This simulation models a biased coin where the probability of heads is $0.35$.

This demonstrates the Law of Large Numbers: as the number of trials increases, the empirical running proportion of heads stabilizes and converges perfectly to the true theoretical probability of $0.35$.

```python
np.random.seed(42) # For reproducibility
n_values = [10, 100, 1000, 10000, 100000]
p_heads = 0.35

plt.figure(figsize=(10, 6))
for n in n_values:
    # Step 1 & 2: Process and single trial is a coin flip (1=Heads, 0=Tails)
    # Step 3 & 4: Repeat and collect results
    flips = np.random.choice([1, 0], size=n, p=[p_heads, 1-p_heads])
    running_proportion = np.cumsum(flips) / np.arange(1, n + 1)

    # Plotting the running proportion
    plt.plot(np.arange(1, n + 1), running_proportion, label=f'n={n}')

plt.axhline(y=p_heads, color='black', linestyle='--', label='True P(Heads) = 0.35')
plt.xscale('log') # Log scale helps visualize the convergence over large n
plt.xlabel('Number of Flips (log scale)')
plt.ylabel('Running Proportion of Heads')
plt.title('D2a: Law of Large Numbers - Biased Coin Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

**D2b: Bootstrap Conditional Probability**

We estimate the conditional probability $P(B=1 \mid A=1)$, where Event A is "High Access to Electricity" and Event B is "High Life Expectancy".

We use bootstrap sampling (sampling with replacement) to create 50,000 independent samples. This allows our sample to act as a proxy for the population, letting us approximate the true probability reliably without relying on the exact analytical formula.


```python
df_clean = data[['access_to_electricity%', 'life_expectancy_at_birth']].dropna()

# Define Event A (Condition): Access to electricity >= 90%
# Define Event B (Outcome): Life expectancy >= 70 years
df_clean['A'] = (df_clean['access_to_electricity%'] >= 90).astype(int)
df_clean['B'] = (df_clean['life_expectancy_at_birth'] >= 70).astype(int)

# Exact Analytical Value
subset_A1 = df_clean[df_clean['A'] == 1]
exact_prob = subset_A1['B'].mean() # This is Count(B=1 and A=1) / Count(A=1)

# Bootstrap Simulation
num_bootstraps = 50000
bootstrap_estimates = []

for _ in range(num_bootstraps):
    # Sample rows with replacement
    sample = df_clean.sample(n=len(df_clean), replace=True)
    sample_subset_A1 = sample[sample['A'] == 1]

    if len(sample_subset_A1) > 0:
        prob = sample_subset_A1['B'].mean()
        bootstrap_estimates.append(prob)

estimated_prob = np.mean(bootstrap_estimates)

print("\n--- D2b Results ---")
print(f"Exact Analytical P(B=1 | A=1): {exact_prob:.4f}")
print(f"Bootstrap Estimated P(B=1 | A=1): {estimated_prob:.4f}")
```

**D2c: Shared Birth Month Simulation**

This simulates the chance that at least two people in a group of size $k$ share the same birth month. As the group size $k$ increases, the probability rises exponentially. This happens because we only need one matching pair, and the number of possible pairs of people grows rapidly as $k$ increases, quickly overwhelming the 12 available months.

```python
k_values = [2, 5, 8, 12, 20]
simulations = 10000
months = list(range(1, 13)) # Months 1 to 12

print("\n--- D2c Results ---")
for k in k_values:
    matches = 0
    for _ in range(simulations):
        # Assign random birth months to k people
        group_months = np.random.choice(months, size=k, replace=True)
        # If length of unique months is less than k, at least two share a month
        if len(set(group_months)) < k:
            matches += 1
    prob = matches / simulations
    print(f"Simulated chance for k={k}: {prob:.4f}")

# Exact analytical chance for k=5
# P(at least 2 share) = 1 - P(all distinct)
# P(all distinct) = (12/12) * (11/12) * (10/12) * (9/12) * (8/12)
exact_k5 = 1 - (12 * 11 * 10 * 9 * 8) / (12**5)
print(f"Exact analytical chance for k=5: {exact_k5:.4f}")
```

## Task D3: Bayesian Update on Real Data

Prior $P(H)$: The initial baseline probability that a randomly selected country has a High GDP, before we know anything else about it.

Likelihood $P(E \mid H)$: The probability of observing High Life Expectancy given that we already know the country has a High GDP.

Posterior $P(H \mid E)$: Our updated, improved probability that a country has a High GDP after observing the new evidence that it has a High Life Expectancy.

```python
# Drop missing values for the selected variables
df_bayes = data[['GDP_current_US', 'life_expectancy_at_birth']].dropna()

# Event E: High Life Expectancy (>= 70)
# Subgroup H: High GDP (Above median)
median_gdp = df_bayes['GDP_current_US'].median()
df_bayes['E'] = (df_bayes['life_expectancy_at_birth'] >= 70).astype(int)
df_bayes['H'] = (df_bayes['GDP_current_US'] > median_gdp).astype(int)

# Calculate Prior, Likelihoods
prior_H = df_bayes['H'].mean() # P(H)
prob_E_given_H = df_bayes[(df_bayes['H'] == 1)]['E'].mean() # P(E|H)
prob_E_given_not_H = df_bayes[(df_bayes['H'] == 0)]['E'].mean() # P(E|~H)

# Bayes Theorem Calculation: P(H|E) = (P(E|H) * P(H)) / P(E)
# P(E) = (P(E|H) * P(H)) + (P(E|~H) * P(~H))
marginal_E = (prob_E_given_H * prior_H) + (prob_E_given_not_H * (1 - prior_H))
posterior_H_given_E = (prob_E_given_H * prior_H) / marginal_E

print("\n--- D3 Results ---")
print(f"Prior P(H): {prior_H:.4f}")
print(f"Likelihood P(E|H): {prob_E_given_H:.4f}")
print(f"Likelihood P(E|~H): {prob_E_given_not_H:.4f}")
print(f"Posterior P(H|E): {posterior_H_given_E:.4f}")

# Sensitivity Analysis: Change prior from 0.1 to 0.9
priors = np.linspace(0.1, 0.9, 50)
posteriors = []

for p in priors:
    # Re-calculate marginal E and posterior for each assumed prior
    m_E = (prob_E_given_H * p) + (prob_E_given_not_H * (1 - p))
    post = (prob_E_given_H * p) / m_E
    posteriors.append(post)

plt.figure(figsize=(8, 5))
plt.plot(priors, posteriors, marker='o', linestyle='-', color='b')
plt.plot([0.1, 0.9], [0.1, 0.9], linestyle='--', color='gray', label='Baseline (Posterior = Prior)')
plt.xlabel('Assumed Prior Probability P(H)')
plt.ylabel('Resulting Posterior Probability P(H|E)')
plt.title('D3: Sensitivity Analysis of the Posterior to the Prior')
plt.legend()
plt.grid(True)
plt.show()
```

## E1. Normality Assessment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats



#####################################################
#  1. Select two numerical variables: one approximately normal and one clearly non-normal.
#####################################################

#get numerical columns
numeric_cols = data.select_dtypes(include=['number']).columns

# Remove columns with too many missing values or constant values
candidates = []

for col in numeric_cols:
    percentage_null = data[col].isnull().mean()*100
    mdata = data[col].dropna()

    if percentage_null > 10:
        continue  # too small to judge reliably

    if mdata.nunique() <= 10:
        continue  # likely categorical disguised as numeric

    candidates.append(col)

print("Candidates:", candidates)

# Compute Skew & Kurtosis for each candidate col
results = []

for col in candidates:
    x = data[col].dropna()

    if len(x) < 3:
        continue # Skip very small samples

    skew = stats.skew(x)
    kurt = stats.kurtosis(x)

    results.append({
        'Variable': col,
        'Skew': skew,
        'Kurtosis': kurt,
    })

summary = pd.DataFrame(results)
print(summary)

#Categorize approximately normal variable and not normal variable
# Approximately normal candidates
normal_candidates = summary[
    (summary['Skew'].between(-0.5, 0.5)) &
    (summary['Kurtosis'].between(-1, 1))
]

# Clearly non-normal candidates
non_normal_candidates = summary[
    (abs(summary['Skew']) > 1) |
    (summary['Kurtosis'] < -1.5) |
    (summary['Kurtosis'] > 2)
]

print("Likely normal candidates:")
print(normal_candidates)

print("\nClearly non-normal candidates:")
print(non_normal_candidates)

#######################################################
# Approximatly normal var : goverment_effectiveness_estimate
# Not normal var : population
#######################################################
normal_var = 'goverment_effectiveness_estimate'
non_normal_var = 'population'

#######################################################
# 2. For each, produce a histogram with fitted normal PDF, an empirical CDF with theoretical normal CDF, and a Q-Q plot.
#######################################################
def analyze_variable(data, var_name):
    x = data[var_name].dropna()

    # Fit normal distribution
    mu, sigma = stats.norm.fit(x)

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # -------------------------
    # Histogram + Normal PDF
    # -------------------------
    sns.histplot(x, kde=False, stat="density", ax=axes[0])
    xmin, xmax = axes[0].get_xlim()
    x_range = np.linspace(xmin, xmax, 100)
    pdf = stats.norm.pdf(x_range, mu, sigma)
    axes[0].plot(x_range, pdf)
    axes[0].set_title(f"{var_name} - Histogram + Normal PDF")

    # -------------------------
    # Empirical CDF vs Normal CDF
    # -------------------------
    sorted_x = np.sort(x)
    ecdf = np.arange(1, len(x)+1) / len(x)
    axes[1].plot(sorted_x, ecdf, label='Empirical CDF')

    normal_cdf = stats.norm.cdf(sorted_x, mu, sigma)
    axes[1].plot(sorted_x, normal_cdf, linestyle='--', label='Normal CDF')
    axes[1].set_title(f"{var_name} - ECDF vs Normal CDF")
    axes[1].legend()

    # -------------------------
    # Q-Q Plot
    # -------------------------
    stats.probplot(x, dist="norm", plot=axes[2])
    axes[2].set_title(f"{var_name} - Q-Q Plot")

    plt.tight_layout()
    plt.show()

    #######################################################
    # 3. Apply the Shapiro–Wilk test and report the statistic and p-value.
    #######################################################
    stat, p_value = stats.shapiro(x)

    print(f"\nShapiro-Wilk Test for {var_name}")
    print(f"Statistic = {stat:.4f}, p-value = {p_value:.6f}")


analyze_variable(data, normal_var)
analyze_variable(data, non_normal_var)
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Aesthetics
plt.style.use('seaborn-v0_8-whitegrid')
PALETTE = {
    'blue': '#2196F3',
    'orange': '#FF6B35',
    'green': '#4CAF50',
    'purple': '#9C27B0',
    'red': '#F44336',
    'grey': '#607D8B',
}

# Reproducable randomness
rng = np.random.default_rng(seed=42)
```

We selected `CO2_Emissions` as our numerical variable for this task because our univariate analysis in Task B revealed it to be not normal, heavily right-skewed with a skewness coefficient of `9.57` and a kurtosis of `101.73`. The Central Limit Theorem (CLT) can be demonstrated clearest using a population that violates the assumption of normality, so using a distribution that is heavily distorted by extreme values should serve as a rigorous stress test for the theorem. This will showcase how the sampling distribution of the mean converges to a symmetric normal curve as the sample size increases, even if the empirical data is non-normal.

```python
# Choose 'CO2_emissions' based on Task B findings that it is highly right-skewed
df = data.copy()
population_data = df ['CO2_emisions'].dropna().values
POP_MEAN = np.mean(population_data)
POP_SD = np.std(population_data, ddof=0)

# Draw 5000 random samples with replacement
REPETITIONS = 5000

# Compute sample means for n= 2, 5, 15, 30, and 100
SAMPLE_SIZES = [2, 5, 15, 30, 100]
```

We drop the missing values for `CO2_emissions` from the World Bank dataset first so that we get clean data to work with. And as we are required to treat the numerical variable's empirical distribution as the population, we use `ddof=0` to divide by $N$ when calculating the standard deviation of the entire population.

```python
# Draw 5000 independent samples of size n from the empirical array,
# and record the sample mean each time
def simulate_clt_empirical(pop_data, sample_sizes, repetitions, rng):
  results = {}
  for n in sample_sizes:
    samples = rng.choice(pop_data, size=(repetitions, n), replace=True)
    means = samples.mean(axis=1)
    results[n] = means
  return results

# Function call to get array of sample means
clt_results = simulate_clt_empirical(
    pop_data = population_data,
    sample_sizes = SAMPLE_SIZES,
    repetitions = REPETITIONS,
    rng = rng
)
```

We are required to draw 5000 samples with replacement, so we use `replace=True` for `rng.choice()` to simulate drawing from an infinitely large population, ensuring the probability of selecting any given value remains constant across all 5000 draws.

```python
# Produce a histogram of sample means with the theoretical CLT normal curve overlaid for each n
fig, axes = plt.subplots(2, 3, figsize=(15,9))
axes_flat = axes.flatten()

for ax, n in zip(axes_flat[:5], SAMPLE_SIZES):
  means = clt_results[n]
  theoretical_sd = POP_SD / np.sqrt(n)

  # Plot empirical histogram
  ax.hist(means, bins=60, density=True, color=PALETTE['blue'],
          alpha=0.65, edgecolor='white', label='Sample means')

  # Overlay theoretical normal curve
  x_plot = np.linspace(means.min(), means.max(), 400)
  clt_pdf = stats.norm.pdf(x_plot, loc=POP_MEAN, scale=theoretical_sd)
  ax.plot(x_plot, clt_pdf, color=PALETTE['red'], lw=2.5,
          label=f'CLT: N(μ, σ²/{n})')

  ax.axvline(POP_MEAN, color='black', ls=':', lw=1.5, alpha=0.7)
  ax.set_title(f'n = {n} | SD of means = {np.std(means):.2f}', fontsize=11)
  ax.set_xlabel('Sample mean x')
  ax.set_ylabel('Density')
  ax.legend(fontsize=8)

# Hide empty subplot
axes_flat[5].set_visible(False)

plt.suptitle(
    'Central Limit Theorem - Empirical World Bank Population (CO2 Emissions) → Normal Sample Means\n'
    f'(Population is extremely right-skewed; yet means converege to a bell curve)',
    fontsize=12, y=1.01
)
plt.tight_layout()
plt.show()
```

To prove CLT visually, we need to compare our empirical data against perfect mathematical theory. So we use `stats.norm.pdf()` to generate an exact bell curve centered on our true population mean, with a spread defined by `theoretical_sd`.

```python
# Create a table comparing empirical and theoretical means and standard deviations of the sampling distribution
print("Comparison of Empirical and Theoretical Means and Standard Deviations of the Sampling Distribution\n")
print(f"{'n':>6} | {'E[X] (empirical)':>22} | {'σ/√n (theoretical)':>22} | {'SD of means (empirical)':>25}")
print("-" * 85)
for n in SAMPLE_SIZES:
  means = clt_results[n]
  theoretical_sd = POP_SD / np.sqrt(n)
  print(f"{n:>6} | {np.mean(means):>22.4f} | {theoretical_sd:>22.4f} | {np.std(means, ddof=0):>25.4f}")
```

**The Central Limit Theorem (CLT)** states that the probability distribution of the sum or average of a large random sample will be roughly normal, regardless of the distribution of the population from which the sample is drawn.

**The term $\sigma/\sqrt{n}$ represents** the theoretical standard error of the mean, which measures the expected distribution of sample means around the true population mean. **Our comparison table verifies the Square Root Law**, showing that as $n$ increases, the empirical standard deviation of the sample means strictly follows the theoretical $\sigma/\sqrt{n}$ calculation, shrinking proportionally to the square root of the sample size.

The underlying population of `CO2_Emissions` is extremely right-skewed. At $n=2$ and $n=5$, the sampling distributions still heavily reflect this original skew. However, visual convergence to a bell curve begins around $n=15$. By $n=30$ and $n=100$, the empirical histograms align closely with the overlaid theoretical normal curves. This visually confirms that the common rule of thumb ($n \ge 30$) should be adequate for `CO2_Emissions` to approximate normality.

## F2. Sample Size Planning

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Compute minumum sample size needed so that 95% confidence interval for proportion has total width <= desired width
def sample_size_for_proportion(desired_total_width: float, p_guess: float = None) -> int:

  margin_of_error = desired_total_width / 2
  z_star = 1.96

  if p_guess is None:
    variance = 0.5 * (1 - 0.5)
  else:
    variance = p_guess * (1 - p_guess)

  n = (z_star**2 * variance) / (margin_of_error**2)
  return int(np.ceil(n))
```

```python
# Derive minimum sample size required for a 95% confidence interval for a proportion
# with total width <= 0.06 using conservative bound p = 0.5
target_width = 0.06
n_conservative = sample_size_for_proportion(desired_total_width = target_width, p_guess = None)
print(f"Required sample size for 95% CI (width={target_width}) with conservative p = 0.5: {n_conservative:,}")

# Calculate proportion of countries above the 70th percentile threshold for Co2 emissions
clean_co2 = df['CO2_emisions'].dropna()
co2_threshold = clean_co2.quantile(0.70)
p_empirical = round((clean_co2 > co2_threshold).mean(), 4)

# Repeat calculation using observed proportion from the dataset as prior estimate
# and compare required sizes (using the calculated empirical proportion)
n_empirical = sample_size_for_proportion(desired_total_width = target_width, p_guess = p_empirical)
print(f"Required sample size for 95% CI (width={target_width}) with empirical p = {p_empirical:,}: {n_empirical:,}")
```

For our empirical proportion value, we are calculated the proportion of countries above the 70th percentile threshold for Co2 emissions, which should give us a stable proportion of around 0.30.

```python
# Plot required sample size as a function of confidence-interval width from 0.02 to 0.20
# for 90%, 95%, and 99% confidence levels.
widths = np.linspace(0.02, 0.20, 200)

# Z-scores for standard confidence levels
Z_90 = 1.645
Z_95 = 1.96
Z_99 = 2.576

# Calculate arrays for plotting using conservative p = 0.5
n_req_90 = np.ceil((Z_90**2 * 0.25) / (widths / 2)**2).astype(int)
n_req_95 = np.ceil((Z_95**2 * 0.25) / (widths / 2)**2).astype(int)
n_req_99 = np.ceil((Z_99**2 * 0.25) / (widths / 2)**2).astype(int)

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(widths, n_req_90, color=PALETTE['green'], lw=2.5, ls='--', label='90% Confidence')
ax.plot(widths, n_req_95, color=PALETTE['orange'], lw=2.5, label='95% Confidence')
ax.plot(widths, n_req_99, color=PALETTE['blue'], lw=2.5, ls='-.', label='99% Confidence')

# Identify approximate elbow point for the 95% level
n_target = int(np.ceil((1.96**2 * 0.25) / (target_width / 2)**2))
ax.scatter(target_width, n_target, color='red', zorder=5)
ax.annotate(f'Elbow/Target (w=0.06, n={n_target})', xy=(target_width, n_target),
            xytext=(target_width + 0.01, n_target + 500), fontsize=10,
            arrowprops=dict(arrowstyle='->', color='red'))

ax.set_title('The Cost of Accuracy: Required Sample Size vs. Desired CI Width', fontsize=12)
ax.set_xlabel('Desired CI total width')
ax.set_ylabel('Minimum sample size n')
ax.legend()
plt.tight_layout()
plt.show()
```

**The factor 1.96** is the critical Z-score ($Z^*$) from the standard normal distribution that captures the central 95 percent of the probability area. Because the normal curve is perfectly symmetric, capturing the middle 95 percent leaves exactly 2.5 percent in the left and right tails. If we use the inverse cumulative distribution function to find the 97.5th percentile, we get a value of around 1.96. As the margin of error is in the denominator and squared in the sample size formula, **the required sample size grows quadratically as the desired interval width shrinks**.

**Using the conservative bound of p=0.5** which maximizes variance for a 95 percent confidence interval with a total width of 0.06, our calculation shows a required sample of **1,068**. Using the empirical proportion of around 0.30 reduces the theoretical variance, lowering the required sample size down to **899**. This demonstrates how prior knowledge of the true proportion can save data collection resources while still maintaining the desired level of precision.

The plotted curves visualize the "cost of accuracy". For the 95 percent confidence level, a clear elbow forms right around our target width of 0.06. Trying to narrow the confidence interval below this point (such as dropping to a width of 0.02) causes the required sample size to spike exponentially, demonstrating severe diminishing returns where **halving the desired width quadruples the required sample size**, making slight gain in precision require massive increases in sample size.
